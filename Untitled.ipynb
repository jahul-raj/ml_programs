```python
import pandas as pd 
import numpy as np
import warnings

warnings.filterwarnings("ignore")
values = np.random.rand(100)
print(values)
```

```python
labels = []
for i in values[:50]:
    if i<=0.5:
        labels.append("class1")
    else:
        labels.append("class2")
labels +=[None]*50
print(labels)
```

```python
data = {
    "point": [f"x{i+1}" for i in range(100)],
    "Values": values,
    "Labels": labels
}
df = pd.DataFrame(data)
df
```

```python
labeled_df=df[df["Labels"].notnull()]
print(labeled_df)
x_train =labeled_df[["Values"]]
y_train =labeled_df[["Labels"]]

unlabeled_df =df[df["Labels"].isna()]
x_test=unlabeled_df[["Values"]]
true_labels=["class1" if x<0.5 else "class2" for x in values[50:]]

from sklearn.neighbors import KNeighborsClassifier as kn
from sklearn.metrics import classification_report


k_values=[1,2,3,4,5,20,30]
for k in k_values:
    knn=kn(n_neighbors=k)
    knn.fit(x_train,y_train)
    predictions = knn.predict(x_test)
    print(f"accuracy for k={k}")
    print(classification_report(true_labels,predictions))
                     

             
```
