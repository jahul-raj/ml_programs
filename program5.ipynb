```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

values=np.random.rand(100)
values
```

```python
labels=[]
for col in values[:50]:
    if col<0.5:
        labels.append('class1')
    else:
        labels.append('class2')
labels +=[None]*50
print(labels)
```

```python
data={'Points':[f"x{i+1}" for i in range(100)],
'Values':values,
      'Labels':labels}
df=pd.DataFrame(data)
df
```

```python
labeled_df=df[df["Labels"].notna()]
x_train=labeled_df[["Values"]]
y_train=labeled_df[["Labels"]]
unlabeled_df=df[df["Labels"].isna()]
x_test=unlabeled_df[["Values"]]
true_labeled=["class1" if x<=0.5 else "class2" for x in values[50:]]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
kvalues=[1,2,3,4,5,20,30]
for i in kvalues:
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    prediction=knn.predict(x_test)
    print(f"Accuracy for k={i}")
    print(classification_report(true_labeled,prediction))
```
