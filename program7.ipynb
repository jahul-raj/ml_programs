```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

# Load data
iris = load_iris(as_frame=True)
df = iris.data
x = df[['sepal length (cm)']]  # Note: Correct column name
y = iris.target

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)
y_pred_linear = linear_reg.predict(x_test)

# Polynomial Regression (degree 2)
poly_model = make_pipeline(PolynomialFeatures(degree=2), StandardScaler(), LinearRegression())
poly_model.fit(x_train, y_train)
y_pred_poly = poly_model.predict(x_test)

# Plot linear regression
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_test, y_pred_linear, color='red', label='Linear Prediction')
plt.title("Linear Regression")
plt.legend()
plt.show()

# Plot polynomial regression
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.scatter(x_test, y_pred_poly, color='green', label='Polynomial Prediction')
plt.title("Polynomial Regression (Degree 2)")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline
import matplotlib.pyplot as plt

# Load data
datas = load_iris(as_frame=True)
df = pd.DataFrame(datas.data)
x = df[['sepal length (cm)']]
y = datas.target

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Linear Regression
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)
y_pred = linear_reg.predict(x_test)

# Polynomial Regression
poly = make_pipeline(PolynomialFeatures(degree=2), StandardScaler(), LinearRegression())
poly.fit(x_train, y_train)

# Smooth line for polynomial
x_line = np.linspace(x.min()[0], x.max()[0], 300).reshape(-1, 1)
y_poly_pred = poly.predict(x_line)

# Plot actual vs linear
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_test, y_pred, color='red', label='Linear Predicted')
plt.legend()
plt.show()

# Plot actual vs polynomial
plt.scatter(x_test, y_test, color='blue', label='Actual')
plt.plot(x_line, y_poly_pred, color='green', label='Polynomial Predicted')
plt.legend()
plt.show()
```
